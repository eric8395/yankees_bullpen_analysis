## Load in 2021 Pitches

```python
# load in all 2021 pitches
pitches_21 = pd.read_csv('all_pitches_2021.csv')
pitches_21.head(3)
```

## Load in 2022 Pitches

```python
# load in all 2022 pitches
pitches_22 = pd.read_csv('all_pitches_2022.csv')
pitches_22.head(3)
```

```python
# define columns to keep
pitching_columns = ['game_year', 'game_date', 'player_name', 'pitch_type', 'stand', 
                    'release_speed', 'release_spin_rate', 'release_extension', 'release_pos_x', 
                    'release_pos_y', 'release_pos_z', 'zone', 'pfx_x', 'pfx_z', 
                    'plate_x', 'plate_z']
```

## Data Dictionary

`release_spin_rate` - Revolutions per minute of pitch as tracked by Statcast.

`release_extension` - Release extension of pitch in feet as tracked by Statcast.

`release_pos_x` - Horizontal Release Position of the ball measured in feet from the catcher's perspective.

`release_pos_y` - Release position of pitch measured in feet from the catcher's perspective.

`release_pos_z` - Vertical Release Position of the ball measured in feet from the catcher's perspective.

`zone` - Zone location of the ball when it crosses the plate from the catcher's perspective.

`pfx_x` - Horizontal movement in feet from the catcher's perspective.

`pfx_z` - Vertical movement in feet from the catcher's perpsective.

`plate_x` - Horizontal position of the ball when it crosses home plate from the catcher's perspective.

`plate_z` - Vertical position of the ball when it crosses home plate from the catcher's perspective.


```python
# redefine dataframe to include only columns to keep
pitches_21 = pitches_21[pitching_columns]
pitches_21.head()
```

```python
# redefine dataframe to include only columns to keep
pitches_22 = pitches_22[pitching_columns]
pitches_22.head()
```

```python
# remove NaN rows
pitches_21 = pitches_21.dropna()
pitches_22 = pitches_22.dropna()

# convert to date time format
pitches_21['game_date'] = pd.to_datetime(pitches_21['game_date'])
pitches_22['game_date'] = pd.to_datetime(pitches_22['game_date'])

# convert pitch movement to inches
pitches_21['pfx_x'] = pitches_21['pfx_x']*12
pitches_21['pfx_z'] = pitches_21['pfx_z']*12

pitches_22['pfx_x'] = pitches_22['pfx_x']*12
pitches_22['pfx_z'] = pitches_22['pfx_z']*12

# convert pitch location to inches
pitches_21['plate_x'] = pitches_21['plate_x']*12
pitches_21['plate_z'] = pitches_21['plate_z']*12

pitches_22['plate_x'] = pitches_22['plate_x']*12
pitches_22['plate_z'] = pitches_22['plate_z']*12
```

```python
# groupby player name and pitch type mean
grouped_pitches_21 = pitches_21.groupby(['player_name','pitch_type']).mean()
grouped_pitches_21 = grouped_pitches_21.reset_index()
```

```python
# groupby player name and pitch type mean
grouped_pitches_22 = pitches_22.groupby(['player_name','pitch_type']).mean()
grouped_pitches_22 = grouped_pitches_22.reset_index()
```

## Combine 2022 & 2021 Average Pitch Metrics

```python
avg_pitch_totals = pd.concat([grouped_pitches_22, grouped_pitches_21])
avg_pitch_totals = avg_pitch_totals.reset_index(drop = True)
avg_pitch_totals
```

```python
# create individual dataframes by specific player
holmes = avg_pitch_totals.loc[avg_pitch_totals['player_name'] == 'Holmes, Clay']

peralta = avg_pitch_totals.loc[avg_pitch_totals['player_name'] == 'Peralta, Wandy']

trivino = avg_pitch_totals.loc[avg_pitch_totals['player_name'] == 'Trivino, Lou']
```

```python
def plot_pitch_movement(data):
    """
    Pass in player specific dataframe and plots the average pitch movement scatter plot for 2021 - 2022 seasons
    """
    
    # define figure size
    sns.set(rc={"figure.figsize":(10, 10)}) 
    sns.set_style('dark')

    # plot scatter plot
    scatter = sns.scatterplot(data = data,
                            x = 'pfx_x',
                            y = 'pfx_z',
                            hue = 'pitch_type', 
                            s = 350,
                            style = 'game_year',
                            palette = 'bright')

    # plot axes
    plt.plot([23, -23], [0, 0], linewidth=3, color = 'black', linestyle='dashed')
    plt.plot([0, 0], [23, -23], linewidth=3, color = 'black', linestyle='dashed')

    # plot title
    scatter.set_title(data['player_name'].unique()[0] +' - Pitch Movement (2021 - 2022) \n' + "(View from Catcher's Perspective)", 
                      weight = "bold", 
                      size = 20)
    
    # Set x-axis label
    plt.xlabel('Horizontal Movement (in)')
    # Set y-axis label
    plt.ylabel('Vertical Movement (in)')

    # set limits of axes
    plt.ylim(-24, 24)
    plt.xlim(-24, 24)

    plt.tight_layout()
    
    # save image
    plt.savefig('images/'+ data['player_name'].unique()[0] +'_pitch_movement.png', bbox_inches = 'tight')
    
    plt.show()
    

    
```

## Release Point

```python
def plot_release_point(data):
    """
    Pass in player specific dataframe and plots the average pitch movement scatter plot for 2021 - 2022 seasons
    """
    
    # define figure size
    sns.set(rc={"figure.figsize":(8, 12)}) 
    sns.set_style('ticks')


    # plot scatter plot
    scatter = sns.scatterplot(data = data,
                            x = 'release_pos_x',
                            y = 'release_pos_z',
                            hue = 'pitch_type', 
                            s = 150,
                            style = 'game_year',
                            palette = 'bright')

    # plot release point
    plt.plot([0, 0], [0.25, max(data['release_pos_z'])], linewidth=2, color = 'black', linestyle = ':')
    plt.plot([0.3, -0.3], [max(data['release_pos_z']), max(data['release_pos_z']) ], linewidth=2, color = 'black', linestyle = ':')
    
    # annotate release point
    plt.annotate('Release Point: ' + str(round(data['release_pos_z'].mean(),2)) + ' ft', 
                 xy=(0, max(data['release_pos_z']) + 1), 
                 xytext=(-0.65, max(data['release_pos_z']) + 0.3), weight = "bold")

    
    # plot plate
    plt.plot([0.708, -.708], [0, 0], linewidth=2, color = 'black')
    plt.plot([0.708, 0.708], [0,-0.708], linewidth=2, color = 'black')
    plt.plot([0.708, 0], [-0.708,-1.417], linewidth=2, color = 'black')
    plt.plot([0, -0.708], [-1.417,-0.708], linewidth=2, color = 'black')
    plt.plot([-0.708, -0.708], [0,-0.708], linewidth=2, color = 'black')

    # plot title
    scatter.set_title(data['player_name'].unique()[0] +' - Average Pitch Release Point (2021 - 2022) \n', 
                      weight = "bold", 
                      size = 18)

    # Set x-axis label
    plt.xlabel('Position X (ft)')
    # Set y-axis label
    plt.ylabel('Position Z (ft)')

    # set limits of axes
    plt.ylim(-2.5, 8)
    plt.xlim(-3, 3)
    
    plt.tight_layout()
    
    # save image
    plt.savefig('images/'+ data['player_name'].unique()[0] +'_release_point.png', bbox_inches = 'tight')

    plt.show()
```

```python
def plot_all_release_point(data, player, date):
    """
    Pass in dataframe, player name (string), and filter date; plots all release points for pitches per filter date
    """
    
    # define figure size
    sns.set(rc={"figure.figsize":(8, 12)}) 
    sns.set_style('ticks')
    
    
    # convert date input into datetime
    from datetime import datetime
    date_format = "%Y-%m-%d"
    date = datetime.strptime(date,date_format)

    
    # plot scatter plot before 2022-7-09
    if date <= (datetime.strptime("2022-7-09", date_format)):
        scatter = sns.scatterplot(data = data[(data['player_name'] == player) 
                                                & (data['game_date'] <= datetime(2022,7,9))], 

                                x = 'release_pos_x',
                                y = 'release_pos_z',
                                hue = 'pitch_type', 
                                s = 15,
                                x_jitter = 200,
                                y_jitter = 200,
                                palette = 'bright')
        
        # plot title after 2022-7-09
        scatter.set_title(player +' - Pitch Release Point (7/9/2022 & Before) \n', 
                          weight = "bold", 
                          size = 18)
        
    # plot after 2022-7-12 
    else:
        scatter = sns.scatterplot(data = data[(data['player_name'] == player) 
                                                & (data['game_date'] > datetime(2022,7,12))], 

                                x = 'release_pos_x',
                                y = 'release_pos_z',
                                hue = 'pitch_type', 
                                s = 15,
                                x_jitter = 200,
                                y_jitter = 200,
                                palette = 'bright')
        
        # plot title after 7/12/22
        scatter.set_title(player +' - Pitch Release Point (After 7/9/2022) \n', 
                          weight = "bold", 
                          size = 18)

    # plot release point
    plt.plot([0, 0], [0.25, max(data['release_pos_z'])], linewidth=2, color = 'black', linestyle = ':')
    plt.plot([0.3, -0.3], [max(data['release_pos_z']), max(data['release_pos_z']) ], linewidth=2, color = 'black', linestyle = ':')
    
    # annotate release point
    plt.annotate('Release Point: ' + str(round(data['release_pos_z'].mean(),2)) + ' ft', 
                 xy=(0, max(data['release_pos_z']) + 1), 
                 xytext=(-0.65, max(data['release_pos_z']) + 0.3), weight = "bold")

    
    # plot plate
    plt.plot([0.708, -.708], [0, 0], linewidth=2, color = 'black')
    plt.plot([0.708, 0.708], [0,-0.708], linewidth=2, color = 'black')
    plt.plot([0.708, 0], [-0.708,-1.417], linewidth=2, color = 'black')
    plt.plot([0, -0.708], [-1.417,-0.708], linewidth=2, color = 'black')
    plt.plot([-0.708, -0.708], [0,-0.708], linewidth=2, color = 'black')

    # Set x-axis label
    plt.xlabel('Position X (ft)')
    # Set y-axis label
    plt.ylabel('Position Z (ft)')

    # set limits of axes
    plt.ylim(-2.5, 8)
    plt.xlim(-3, 3)
    
    plt.tight_layout()
    
    # save image
    plt.savefig('images/'+ player + str(date) + '_release_points.png' , bbox_inches = 'tight')

    plt.show()
```

## Spinrate

```python
def plot_spinrate(data, player):
    """
    Plot a lineplot of spinrate for pitches over time, input pitches df and name of player (string)
    """
    # define figure size
    sns.set(rc={"figure.figsize":(15, 6)}) 
    sns.set_style('darkgrid')
    
        
    # assigning colors to each label
    palette = {'FF': 'red', 'CU':'blue', 'SI':'green', 'SL':'orange',
               'CH': 'purple', 'FC': 'brown'}
    

    # plot line plot of spin rate by game date
    spinrate_lineplot = sns.lineplot(
                                        data = data[data['player_name'] == player], 
                                        x = 'game_date', 
                                        y = 'release_spin_rate', 
                                        hue = 'pitch_type', ci = 0, palette = palette
                                        )
    
    # date where MLB started checking pitchers
    date = datetime(2021, 6, 22)

    # plot vertical line of date cut off
    plt.axvline(date, 
                color = 'black', 
                label = 'Foreign Substance Enforcement (6/22/21)', 
                linestyle = 'dashed')

    # plot legend
    plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0.25)

    # plot title
    spinrate_lineplot.set_title(player +' - Pitch Spinrates (2021) \n', 
                      weight = "bold", 
                      size = 16)

    # Set x-axis label
    plt.xlabel('Date')
    # Set y-axis label
    plt.ylabel('Release Spinrate (RPM)')

    # rotate x labels
    plt.xticks(rotation=45)
    
    plt.tight_layout()
    # save image
    plt.savefig('images/'+ player +'_2021_spinrates.png', bbox_inches = 'tight')
    
    plt.show()
```

## Pitch Location by Year

```python
def plot_pitch_location(data_1, data_2, player):
    """
    Pass in dataframe, player name (string) and plots all location of pitches through strike zone
    """
    
    # define figure size
    sns.set(rc={"figure.figsize":(12, 7)}) 
    sns.set_style('darkgrid')
    
    # two subplots 
    fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, sharex=True)
    
    # assigning colors to each label
    palette = {'FF': 'red', 'CU':'blue', 'SI':'green', 'SL':'orange',
               'CH': 'purple', 'FC': 'brown'}
    
    # plot scatter plot 2021
    scatter_1 = sns.scatterplot(data = data_1[(data_1['player_name'] == player)], 
                                x = 'plate_x',
                                y = 'plate_z',
                                hue = 'pitch_type', 
                                s = 50,
                                x_jitter = 200,
                                y_jitter = 200,
                                alpha = 0.7,
                                palette = palette, ax = ax1)
    
    # plot scatter plot 2022
    scatter_2 = sns.scatterplot(data = data_2[(data_2['player_name'] == player)], 
                                x = 'plate_x',
                                y = 'plate_z',
                                hue = 'pitch_type', 
                                s = 50,
                                x_jitter = 200,
                                y_jitter = 200,
                                alpha = 0.7,
                                palette = palette, ax = ax2)

    # plot title 1
    scatter_1.set_title(str(data_1['game_year'][0]), 
                      weight = "bold", 
                      size = 14)
    # plot title 2
    scatter_2.set_title(str(data_2['game_year'][0]), 
                      weight = "bold", 
                      size = 14)
    
    # plot strike zone
    ax1.plot([-8.5, 8.5], [16, 16], linewidth=3, color = 'black', linestyle = 'dashed', dashes = (5,3))
    ax1.plot([-8.5, -8.5], [42, 16], linewidth=3, color = 'black', linestyle = 'dashed', dashes = (5,3))
    ax1.plot([-8.5, 8.5], [42, 42], linewidth=3, color = 'black', linestyle = 'dashed', dashes = (5,3))
    ax1.plot([8.5, 8.5], [42, 16], linewidth=3, color = 'black', linestyle = 'dashed', dashes = (5,3))
    
    # plot strike zone
    ax2.plot([-8.5, 8.5], [16, 16], linewidth=3, color = 'black', linestyle = 'dashed', dashes = (5,3))
    ax2.plot([-8.5, -8.5], [42, 16], linewidth=3, color = 'black', linestyle = 'dashed', dashes = (5,3))
    ax2.plot([-8.5, 8.5], [42, 42], linewidth=3, color = 'black', linestyle = 'dashed', dashes = (5,3))
    ax2.plot([8.5, 8.5], [42, 16], linewidth=3, color = 'black', linestyle = 'dashed', dashes = (5,3))

    
    # set title 
    fig.text(0.5, 1, player + " - Pitch Location", ha = 'center', weight = 'bold', size = 18)

    # Set x-axis label
    fig.text(0.5, 0, "Horizontal Position Across Homeplate (in)", ha = 'center', weight = 'bold', size = 14)
    
    # Set y-axis label
    fig.text(0, 0.3, "Vertical Position Across Homeplate (in)", ha = 'center', weight = 'bold', size = 14, rotation = 90)

    # set limits of axes
    plt.xlim(-36, 36)
    plt.ylim(-20, 84)
    
    plt.tight_layout()
    plt.show()
```

```python
plot_pitch_location(pitches_21, pitches_22, 'Holmes, Clay')
```

```python
plot_pitch_location(pitches_21, pitches_22, 'Peralta, Wandy')
```

## Pitch Location by Batter

```python
def plot_pitch_location_by_batter(data_1, data_2, player):
    """
    Pass in dataframe, player name (string) and plots all location of pitches through strike zone 
    by year and batter stance
    """
    
    # define figure size
    sns.set(rc={"figure.figsize":(16, 16)}) 
    sns.set_style('darkgrid')
    
    # four subplots 
    fig, ax = plt.subplots(2, 2, sharey=True, sharex=True)
    
    # assigning colors to each label
    palette = {'FF': 'red', 'CU':'blue', 'SI':'green', 'SL':'orange',
               'CH': 'purple', 'FC': 'brown'}
    
    # plot scatter plot 2021 - RIGHT
    scatter_1 = sns.scatterplot(data = data_1[(data_1['player_name'] == player) 
                                              & (data_1['stand'] == 'R')], 
                                x = 'plate_x',
                                y = 'plate_z',
                                hue = 'pitch_type', 
                                s = 70,
                                x_jitter = 200,
                                y_jitter = 200,
                                alpha = 0.7,
                                palette = palette, ax = ax[0,0]) 
    
    # plot scatter plot 2021 - LEFT
    scatter_2 = sns.scatterplot(data = data_1[(data_1['player_name'] == player) 
                                              & (data_1['stand'] == 'L')], 
                                x = 'plate_x',
                                y = 'plate_z',
                                hue = 'pitch_type', 
                                s = 70,
                                x_jitter = 200,
                                y_jitter = 200,
                                alpha = 0.7,
                                palette = palette, ax = ax[0,1])
    
    # plot scatter plot 2022 - RIGHT
    scatter_3 = sns.scatterplot(data = data_2[(data_2['player_name'] == player)  
                                             & (data_2['stand'] == 'R')], 
                                x = 'plate_x',
                                y = 'plate_z',
                                hue = 'pitch_type', 
                                s = 70,
                                x_jitter = 200,
                                y_jitter = 200,
                                alpha = 0.7,
                                palette = palette, ax = ax[1,0])
    
    # plot scatter plot 2022 - LEFT
    scatter_4 = sns.scatterplot(data = data_2[(data_2['player_name'] == player)  
                                             & (data_2['stand'] == 'L')], 
                                x = 'plate_x',
                                y = 'plate_z',
                                hue = 'pitch_type', 
                                s = 70,
                                x_jitter = 200,
                                y_jitter = 200,
                                alpha = 0.7,
                                palette = palette, ax = ax[1,1])

    # plot title 1
    scatter_1.set_title('Right-Handed Batters', 
                      weight = "bold", 
                      size = 20)
    # plot title 2
    scatter_2.set_title('Left-Handed Batters', 
                      weight = "bold", 
                      size = 20)
    
    # plot strike zones
    ax[0,0].plot([-8.5, 8.5], [16, 16], linewidth=3, color = 'black', linestyle = 'dashed', dashes = (5,3))
    ax[0,0].plot([-8.5, -8.5], [42, 16], linewidth=3, color = 'black', linestyle = 'dashed', dashes = (5,3))
    ax[0,0].plot([-8.5, 8.5], [42, 42], linewidth=3, color = 'black', linestyle = 'dashed', dashes = (5,3))
    ax[0,0].plot([8.5, 8.5], [42, 16], linewidth=3, color = 'black', linestyle = 'dashed', dashes = (5,3))
    
    ax[0,1].plot([-8.5, 8.5], [16, 16], linewidth=3, color = 'black', linestyle = 'dashed', dashes = (5,3))
    ax[0,1].plot([-8.5, -8.5], [42, 16], linewidth=3, color = 'black', linestyle = 'dashed', dashes = (5,3))
    ax[0,1].plot([-8.5, 8.5], [42, 42], linewidth=3, color = 'black', linestyle = 'dashed', dashes = (5,3))
    ax[0,1].plot([8.5, 8.5], [42, 16], linewidth=3, color = 'black', linestyle = 'dashed', dashes = (5,3))

    ax[1,0].plot([-8.5, 8.5], [16, 16], linewidth=3, color = 'black', linestyle = 'dashed', dashes = (5,3))
    ax[1,0].plot([-8.5, -8.5], [42, 16], linewidth=3, color = 'black', linestyle = 'dashed', dashes = (5,3))
    ax[1,0].plot([-8.5, 8.5], [42, 42], linewidth=3, color = 'black', linestyle = 'dashed', dashes = (5,3))
    ax[1,0].plot([8.5, 8.5], [42, 16], linewidth=3, color = 'black', linestyle = 'dashed', dashes = (5,3))
    
    ax[1,1].plot([-8.5, 8.5], [16, 16], linewidth=3, color = 'black', linestyle = 'dashed', dashes = (5,3))
    ax[1,1].plot([-8.5, -8.5], [42, 16], linewidth=3, color = 'black', linestyle = 'dashed', dashes = (5,3))
    ax[1,1].plot([-8.5, 8.5], [42, 42], linewidth=3, color = 'black', linestyle = 'dashed', dashes = (5,3))
    ax[1,1].plot([8.5, 8.5], [42, 16], linewidth=3, color = 'black', linestyle = 'dashed', dashes = (5,3))
    
    # set title 
    fig.text(0.5, 1, player + " - Pitch Location", ha = 'center', weight = 'bold', size = 28)

    # Set x-axis label
    fig.text(0.5, 0, "Horizontal Position Across Homeplate (in)", ha = 'center', weight = 'bold', size = 16)
    
    # Set y-axis label
    fig.text(0, 0.4, "Vertical Position Across Homeplate (in)", ha = 'center', weight = 'bold', size = 16, rotation = 90)
    
    # Set year labels
    fig.text(1.02, 0.73, "2021", ha = 'center', weight = 'bold', size = 20)
    fig.text(1.02, 0.25, "2022", ha = 'center', weight = 'bold', size = 20)

    # set limits of axes
    plt.xlim(-36, 36)
    plt.ylim(-20, 84)
    
    plt.tight_layout()
    plt.show()
    
    # save image
    fig.savefig('images/'+ player +'pitch_locations_by_batter.png', dpi = 300, bbox_inches = 'tight')
    
```
