```python
# load in all 2021 pitches
pitches_21 = pd.read_csv('all_pitches_2021.csv')
pitches_21.head(3)
```

```python
# define columns to keep
pitching_columns = ['game_date', 'player_name', 'pitch_type', 'release_speed', 
                    'release_spin_rate', 'release_extension', 'release_pos_x', 
                    'release_pos_y', 'release_pos_z', 'zone', 'pfx_x', 'pfx_z', 
                    'plate_x', 'plate_z']
```

**Data Dictionary:**

`release_spin_rate` - Revolutions per minute of pitch as tracked by Statcast.

`release_extension` - Release extension of pitch in feet as tracked by Statcast.

`release_pos_x` - Horizontal Release Position of the ball measured in feet from the catcher's perspective.

`release_pos_y` - Release position of pitch measured in feet from the catcher's perspective.

`release_pos_z` - Vertical Release Position of the ball measured in feet from the catcher's perspective.

`zone` - Zone location of the ball when it crosses the plate from the catcher's perspective.

`pfx_x` - Horizontal movement in feet from the catcher's perspective.

`pfx_z` - Vertical movement in feet from the catcher's perpsective.

`plate_x` - Horizontal position of the ball when it crosses home plate from the catcher's perspective.

`plate_z` - Vertical position of the ball when it crosses home plate from the catcher's perspective.


```python
# redefine dataframe to include only columns to keep
pitches_21 = pitches_21[pitching_columns]
pitches_21.head()
```

```python
# remove NaN rows
pitches_21 = pitches_21.dropna()
```

```python
# convert pitch movement to inches
pitches_21['pfx_x'] = pitches_21['pfx_x']/12
pitches_21['pfx_z'] = pitches_21['pfx_z']/12
```

```python
# groupby player name and pitch type mean
grouped_pitches_21 = pitches_21.groupby(['player_name','pitch_type']).mean()

grouped_pitches_21
```

```python
holmes_21 = grouped_pitches_21.iloc[grouped_pitches_21.index.get_level_values('player_name') == 'Holmes, Clay']
peralta_21 = grouped_pitches_21.iloc[grouped_pitches_21.index.get_level_values('player_name') == 'Peralta, Wandy']
trivino_21 = grouped_pitches_21.iloc[grouped_pitches_21.index.get_level_values('player_name') == 'Trivino, Lou']
```
